```python
class StudentDatabase :
    def __init__(self,name,id):
        self.name=name
        self.id=id
        self.grades={}

    def calculateGPA(self,infos,semester):
        self.semester=semester
        self.grades[self.semester]={}
        self.cg=0
        self.course=[]
        sum=0
        for i in infos:
            course,cg= i.split(':')
            sum+=3*float(cg)
            self.course.append(course)
        totalcg=sum/(3*len(infos))
        self.cg=totalcg
        self.grades[self.semester][tuple(self.course)]=self.cg
    def printDetails(self):
        print("Name:", self.name)
        print("ID:", self.id)
        for semester,courses in self.grades.items():
            print(f"Courses taken in {semester}:")
            for tup,cg in courses.items():
                for i in tup:
                    print(i)
                print(cg)

# Write your code here
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()

```

```python
class Department:
    def __init__(self,name='ChE',sections=5):
        self.name=name
        self.sections=sections
        print(f'The {self.name} department has {self.sections} sections')
        self.avg=0
    def add_students(self,*st):
        total=0
        if len(st)==self.sections:
            for i in st:
                total+=int(i)
            self.avg=total/self.sections
            print(f"THe {self.name} department has an average of {self.avg} students in each sections")
        else:
            print(f"The {self.name} department doesn't have {len(st)} sections ")

    def merge_Department(self,*s1):
        sum=0
        for i in s1:
            print(f'{i.name} is merged to {self.name}.')
            sum+=(i.avg*i.sections)
        self.avg=((self.avg*self.sections)+sum)/(self.sections)
        return f'Now the {self.name} has an average of {self.avg:.2f} students in each section.'

d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify:
    def __init__(self,name='Guest'):
        self.n = name
        if self.n != 'Guest':
            print(f'Welcome {self.n} to Shopidify')
        else:
            print('Welcome to Shopidify')
        self.d = {}

    def add_to_cart(self,item,amount=1):
        if type(item) == str:
            if item not in self.d:
                self.d[item] = amount
            else:
                self.d[item] += amount

        elif type(item) == list:
            for i in range(len(item) - 1):
                if type(item[i]) == str and type(item[i + 1]) == int:
                    if item[i] not in self.d:
                        self.d[item[i]] = item[i + 1]
                    else:
                        self.d[item[i]] += item[i + 1]

    def display_cart(self):
        print(f'Items in the cart for {self.n}')
        for i,j in self.d.items():
            print(f"- {i}: {j}X")

    def display_history(self):
        print(f"Purchase history for {self.n}")
        print(f"Transaction 1:")
        for i,j in self.d.items():
            print(f"- {i}: {j}X")

    def checkout(self):
        print(f'Checkout completed for {self.n}')
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")

```

```python
class Shape:

  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base

  def get_height_base(self):
    return "Height: "+str(self.height)+", Base: "+str(self.base)

class triangle (Shape):
    def __init__(self,name='Default',height=0,base=0):
        super().__init__(name,height,base)
    def calcArea(self):
        self.area=(0.5)*(self.height)*(self.base)
    def printDetail(self):
        x = super().get_height_base()
        y = f"Shape name: {self.name}\n{x}, Area: {float(self.area)}"
        return y

class trapezoid(Shape):
  def __init__(self,name,height,base,side):
     super().__init__(name,height,base)
     self.side = side

  def calcArea(self):
    self.area = 0.5*self.height*(self.base+self.side)

  def printDetail(self):
        x = super().get_height_base()
        y = f"Shape name: {self.name}\n{x}, Side_A: {self.side}, Area: {float(self.area)}"
        return y

tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.account_type = "General"
        self.maturity = 0

    def print_details(self):
        print("------ Account details ------")
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
        print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds.")

    def year_passed(self, year):
        self.maturity += year
        print(f"Maturity of the account: {self.maturity} years")


class SavingsAccount(Account):
    def __init__(self, account_number, balance, interest_rate, minimum_limit):
        super().__init__(account_number, balance)
        self.account_type = "Savings"
        self.interest_rate = interest_rate
        self.minimum_limit = minimum_limit

    def print_details(self):
        super().print_details()
        print(f'Interest Rate: {self.interest_rate}')
        print(f'Minimum Limit: {self.minimum_limit}')


    def withdraw(self, amount):
        if self.balance >= amount:
            if amount<=self.minimum_limit:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
            else:
                print("Insufficient funds.")
        else:
                print("Insufficient funds.")



    def apply_interest(self):
        interest_amount = self.balance * self.interest_rate
        self.balance += interest_amount
        print(f'Interest applied new Balance {self.balance}')


class FixedDepositAccount(Account):
    def __init__(self, account_number, balance, maturity_period):
        super().__init__(account_number, balance)
        self.account_type = "Fixed Deposit"
        self.maturity_period = maturity_period

    def withdraw(self, amount):
        if self.maturity >= self.maturity_period:
            super().withdraw(amount)
        else:
            print("Can not withdraw, Account is not matured")

    def year_passed(self, year):
        super().year_passed(year)

# Driver Code
print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("SA123", 1000, 0.05, 800)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1 = FixedDepositAccount("FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```

```python
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self,marks,time,*subj):
        super().__init__(marks)
        self.time=time
        self.subj=list(subj)
        self.parts=len(self.subj)+2

    def examSyllabus(self):
        x = super().examSyllabus()
        return x + ", " + ', '.join(self.subj)
    def examParts(self):
        x=super().examParts()
        for i in range (len(self.subj)):
            if i==0:
                x +=f'Part {i+3} - {self.subj[i]}'
            else:
                 x += f'\nPart {i+3} - {self.subj[i]}'
        return x
    def __str__(self):
        return f"Marks {self.marks} Time {self.time} Number of Parts {self.parts}"

engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
class Transport:                                                    #Good One
    total_traveller = 0

    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare

    def __str__(self):
        s =  "Name: "+self.name+", Base fare: "+str(self.baseFare)
        return s

class Bus(Transport):
    def __init__(self,place,fare):
        super().__init__(place,fare)
        print(f'Base fare of volvo is {fare} ')
        self.dict1={}


    def addPassengerWithBags(self, *info):
        fare= self.baseFare
        for j in range(0, len(info), 2):
            if 3 <= info[j + 1] <= 5:
                fare += 60
            elif info[j + 1] > 5:
                fare += 105
            else:
                fare = self.baseFare
            self.dict1[info[j]] = fare
        Transport.total_traveller+=len(self.dict1)

    def __str__(self):
            x = super().__str__()
            x += f'\nTotal Passengers: {len(self.dict1)}\n'
            x += f'Passenger Details:\n'
            for passenger, fare in self.dict1.items():
                x += f'Name: {passenger}, Fare: {fare}\n'
            return x
class Train(Bus):
    def __init__(self, name, fare):
        super().__init__(name, fare)


t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6,  "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller)
```

```python
class Shape:

  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base

  def get_height_base(self):
    return "Height: "+str(self.height)+",Base: "+str(self.base)



class triangle(Shape):
    def __init__(self,name='Defualt',height=0,base=0):
        super().__init__(name,height,base)

    def calcArea(self):
        self.area = 0.5*self.height*self.base

    def printDetail(self):
        print(f"Shape name: {self.name}")
        st = f"{self.get_height_base()}\nArea: {float(self.area)}"
        return st

class trapezoid(Shape):
  def __init__(self,name,height,base,side):
     super().__init__(name,height,base)
     self.side = side

  def calcArea(self):
    self.area = 0.5*self.height*(self.base+self.side)

  def printDetail(self):
    print(f"Shape name: {self.name}")
    st =f"{self.get_height_base()}, Side_A: {self.side}\nArea: {float(self.area)}"
    return st




tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
class Workspace:
    def __init__(self, workspaceName, type):
        self.workspaceName = workspaceName
        self.type = type
        self.channels = {}
    def checkChannelExistence(self, channelName):
        if channelName in self.channels:
            return True
            return False
    def __str__(self):
        return f"Name: {self.workspaceName}\nType:{self.type}\n"

class  SlackWorkspace(Workspace):
    total_workspaces=0
    def __init__(self,workspaceName,type,sections=3):
        super().__init__(workspaceName,type)
        self.sections=sections
        print(f"CSE111Workspace is created.")
        SlackWorkspace.total_workspaces+=1
        self.channels={}
        self.total_members=0

    def createChannels(self):
        for i in range(1,self.sections+1):
            self.channels[f"Section-{str(i)}"]=[]

    def addMembersToChannel(self,*mem):
        for i in mem:
            name,num=i.split('_')
            x=f"Section-{str(num)}"
            if x in self.channels:
                self.channels[x]=[name]
                self.total_members+=1
            else:
                print(f"Sorry {name} section-{num} channel not found in {self.workspaceName}")

    def __str__(self):
        x=super().__str__()
        return f"{x}Total Members {self.total_members}\nTotal Channels {len(self.channels)}\nChannel Info : {self.channels}"

print(f"Total no. of workspaces:{SlackWorkspace.total_workspaces}")
print("1=============================")
CSE111Workspace = SlackWorkspace("CSE111Workspace",
"Education",5)
print("2=============================")
print(CSE111Workspace)
print(f"Total no. of workspaces:{SlackWorkspace.total_workspaces}")
print("3=============================")
CSE111Workspace.createChannels()
print("4=============================")
print(CSE111Workspace)
print("5=============================")
john, harry, adil = "John_2", "Harry_10", "Adil_1"
alice, bob = "Alice_2","Bob_3"
print("6=============================")
CSE111Workspace.addMembersToChannel(john, harry)
CSE111Workspace.addMembersToChannel(adil)
print("7=============================")
del(CSE111Workspace.channels["Section-3"])
print(CSE111Workspace)
print("8=============================")
CSE111Workspace.addMembersToChannel(alice, bob)
print("9=============================")
print(CSE111Workspace)
print("10=============================")
CSE320Workspace = SlackWorkspace("CSE320Workspace","Education")
CSE320Workspace.createChannels()
print("11=============================")
print(CSE320Workspace)
print("12=============================")
print(f"Total no. of workspaces:{SlackWorkspace.total_workspaces}")
```
